```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data(file_path: str) -> pd.DataFrame:
    data = pd.read_csv(file_path, index_col=0)
    return data
```

```python
raw_data = load_data('data/raw/BIT_AI_assignment_data.csv')
raw_data = raw_data.drop(columns=['Machine ID', 'Sales ID'])
print(raw_data.shape)
```

```python
print(raw_data.head())
```

```python
print("Variable names:", raw_data.columns.tolist())
TOTAL_TO_REMOVE = ['Machine ID', 'Sales ID']
```

```python
total_missing_values = raw_data.isnull().sum().sum()
print("Total missing values in dataset (%):", total_missing_values / (raw_data.shape[0] * raw_data.shape[1]))
print("Number of columns with missing values:", raw_data.isnull().any().sum())
print("Names of columns without missing values:", raw_data.columns[raw_data.isnull().sum() == 0].tolist())
print("Names of columns with missing values:", raw_data.columns[raw_data.isnull().sum() > 0].tolist())
null_values_percentages = raw_data.isnull().sum()/412698 
null_values_nonzero = null_values_percentages[null_values_percentages > 0]
print("Missing values per column (ranked) (%):\n", null_values_nonzero.sort_values(ascending=False))
```

```python
# drop if >90% missing, confirmed to reduce r^2 by only 0.3%
cols_to_drop = null_values_nonzero[null_values_nonzero > 0.9].index.tolist()
raw_data = raw_data.drop(columns=cols_to_drop)
TOTAL_TO_REMOVE += cols_to_drop
```

```python
categorical = raw_data.select_dtypes(include=['object']).shape[1]
numerical = raw_data.select_dtypes(include=[np.number]).shape[1]
print("Number of categorial variables:", categorical)
print("Number of numerical variables:", numerical)
print("Rest", raw_data.shape[1] - categorical - numerical)
```

```python
# Compute statistics per variable
stats = raw_data.describe().T
print(stats)
```

```python
to_remove_std = stats[stats['std'] == 0].index.tolist()
print("Variables with zero standard deviation (to remove):", to_remove_std)

nunique_counts = raw_data.nunique()
to_remove_unique = nunique_counts[nunique_counts == 1].index.tolist()
print("Variables with a single unique value (to remove):", to_remove_unique)
print("Variables with unique values count:\n", nunique_counts.sort_values())
```

```python
# make a histogram for years
plt.hist(raw_data['Year Made'], edgecolor='black')
plt.title('Distribution of Vehicles by Year')
plt.xlabel('Year')
plt.ylabel('Number of Vehicles')
```

```python
year_1000_count = (raw_data['Year Made'] == 1000).sum()
above_1900_count = (raw_data['Year Made'] > 1900).sum()
above_2025_count = (raw_data['Year Made'] > 2025).sum()
print("Number of vehicles with 'Year Made' > 1900:", above_1900_count)
print("Number of vehicles with 'Year Made' = 1000:", year_1000_count)
print("Number of vehicles with 'Year Made' > 2025:", above_2025_count)
print("Remaing years", raw_data.shape[0] - above_1900_count - year_1000_count)
```

```python
# It looks like we need to clean the year made column of the 1000 entries, replace them with NaN
raw_data['Year Made'] = raw_data['Year Made'].replace(1000, np.nan)
```

```python
# make a histogram for years
plt.hist(raw_data['Year Made'], edgecolor='black')
plt.title('Distribution of Vehicles by Year')
plt.xlabel('Year')
plt.ylabel('Number of Vehicles')
```

```python
# do some processing on the dates as they seem important but ill-represented
raw_data['Sales date'] = pd.to_datetime(raw_data['Sales date'])
raw_data['SalesYear'] = raw_data['Sales date'].dt.year
raw_data['AgeOfMachine'] = raw_data['SalesYear'] - raw_data['Year Made']
raw_data = raw_data.drop(columns=['Sales date'])
TOTAL_TO_REMOVE += ['Sales date']
TOTAL_ADDED = ['SalesYear', 'AgeOfMachine']
```

```python
plt.figure(figsize=(12, 6))
plt.hist(raw_data['Sales Price'], bins=50)
plt.title('Distribution of Sales Price')
plt.xlabel('Sales Price')
plt.ylabel('Frequency')
plt.savefig('sales_price_distribution.png')

plt.figure(figsize=(12, 6))
plt.hist(np.log1p(raw_data['Sales Price']), bins=50)
plt.title('Distribution of log(Sales Price + 1)')
plt.xlabel('log(Sales Price + 1)')
plt.ylabel('Frequency')
plt.savefig('log_sales_price_distribution.png')
```

```python
# as the log labels look better, process them in the dataset
raw_data['Sales Price'] = np.log1p(raw_data['Sales Price'])
```

```python
print(TOTAL_TO_REMOVE, TOTAL_ADDED)
print(raw_data.shape)
```

```python
# save the new dataset in data/processed
raw_data.to_csv('data/processed/processed_v1.csv', index=False)
```
